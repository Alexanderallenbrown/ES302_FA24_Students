## Transformation Functions

````python
import numpy as np

def create2DRotation(theta):
    """ Create a 2D rotation matrix BA_R (R from B to A) to express the unit vectors of coordinate system A
        in terms of the unit vectors of coordinate system A, which is not.
        If you wish to do the opposite (transform unit vectors of A into B's coordinates), you can invert this matrix.
        The angle you feed this function should be in RADIANS.

        note that the matrix equation here is [xhata,yhata]' = R*[xhatb,yhatb]' where ' represents transpose.
    """
    return np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])

def create2DTransform(aX_b,aY_b,theta):
    """ Create a 2D transform matrix that takes any point in coordinate system B and expresses it in A's coordinate system.
        This performs a translation and a rotation assuming that coordinate system B's position in A's coordinates is known,
        and the angle theta by which coordinate system B is rotated relative to A is known.
    """
    #create rotation matrix
    R = create2DRotation(theta)
    #create vector expressing B's position relative to A's origin in A coordinates
    aPb = np.array([[aX_b],[aY_b]])
    #stack these two horizontally
    baT = np.hstack((R,aPb))
    #now stack this vertically with a [0,0,1] to turn into homogeneous transformation
    baT = np.vstack((baT,np.array([0,0,1])))
    #now return this transformation
    return baT

def transform2DBA(bXq,bYq,T):
    """transform a point q expressed in coordinate system B's frame into coordinate system A's frame, where
        T represents the homogeneous transformation from B coords to A coords, assuming that coordinate frame
        B is rotated by a known angle theta relative to A, and is located at a known offset in A's frame.
    """
    result = np.dot(T,np.array([[bXq],[bYq],[1]]))
    #returns aXq,aYq (q's X and Y coordinate in A's frame)
    return result[0][0],result[1][0]

def transform2DAB(aXq,aYq,T):
    """transform a point q expressed in coordinate system A's frame into coordinate system B's frame, where
        where T represents the homogeneous transformation from B coords to A coords, assuming that coordinate frame
        B is rotated by a known angle theta relative to A, and is located at a known offset in A's frame.
        This function uses the INVERSE of T to accomplish this.
    """
    result = np.dot(np.linalg.inv(T),np.array([[aXq],[aYq],[1]]))
    #returns bXq,bYq (q's X and Y coordinate in B's frame)
    return result[0][0],result[1][0]
````

````python
#prerequisite libraries for working with TriP kinematics library
from math import radians
import numpy as np

#solver class built in to trip
from trip_kinematics.Solver import SimpleInvKinSolver
#kinematic group class (used for constructing robots)
from trip_kinematics.KinematicGroup import KinematicGroup
#transformation class (used for constructing transformations)
from trip_kinematics.Transformation import Transformation
#robot class from trip. Renamed here to avoid conflics with WeBots
from trip_kinematics.Robot import Robot as TriPRobot #do this to avoid confusion with Webots Robot class
#import forward kinematics from the Robot superclass
from trip_kinematics.Robot import forward_kinematics


"""our first task is to come up with the frames for ArtyBot.
We will at first ONLY consider the XY plane motion of Artybot. This allows us
to familiarize ourselves with the trip_kinematics library in 2D, where we have
a known analytical solution to compare to.
"""

#properties of the Artybot robot
l1 = .085
l2 = .053
th1_static = 18*3.1415/180.0#.38 #radians. tan(40mm/100mm)
th2_static = 18*3.1415/180.0

#now we develop the transformation between the base frame A (allocentric)
#and the link 1 frame (B). There is no translation here.
#it is possible to indicate a rotation to correct for the offset between servo angle and
#'true' theta_1 as described in our derivation, but we will not.
AB_trans = Transformation(name='AB_trans', values={'tx': 0, 'ty':0, 'tz': 0})
#now develop the joint between frame A and frame B
L1_joint = Transformation(name='L1_joint',
                             values={'rz': 0},
                             state_variables=['rz'],
                             parent=AB_trans)
#now develop the transformation between link 1 (frame B) and link 2 (frame C)
BC_trans = Transformation(name='BC',values={'tx':l1,'ty':0,'tz':0},parent=L1_joint)
#now develop the joint between B and C
L2_joint = Transformation(name='L2_joint',
                             values={'rz': 0},
                             state_variables=['rz'],
                             parent=BC_trans)
#now develop the transformation between C and end effector
Cpen_trans = Transformation(name='Cpen_trans',values={'tx':l2,'ty':0,'tz':0},parent=L2_joint)

#conbine these transforms into a TriP robot object
arty_object = TriPRobot([AB_trans,L1_joint,BC_trans,L2_joint,Cpen_trans])

#now we can use arty!
print("things inside of the Arty Robot")
print(dir(arty_object))
print("actuated state of Arty:")
print(arty_object.get_actuated_state())
print("forward kinematics for this state:")
#trip_kinematics.Robot.forward_kinematics(robot: trip_kinematics.Robot.Robot, endeffector)
print(forward_kinematics(arty_object,'Cpen_trans'))

#now we can set the actuated state of the robot if we want
#note the format it wants is a dict.
arty_object.set_actuated_state({'L1_joint_rz':0.1,'L2_joint_rz':0.1})
print("new actuated state:")
print(arty_object.get_actuated_state())
print("new forward kinematics:")
print(forward_kinematics(arty_object,'Cpen_trans'))

#initialize solver. we give it the robot, and the end effector frame.
artyIK = SimpleInvKinSolver(arty_object,'Cpen_trans')

#use solver to do inverse kin.
soln = artyIK.solve_actuated(np.array([.075,.075,0]))
print(soln)
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

#function to update dynamics
def updateMotion(x,u):
    #just a simple model for now
    x = x+u
    return x


#function to get the index j in Bel given an x.
def getCell(x,Bel,d):
    m = np.size(Bel)
    j = int((x-d/(2*m))*m/d)
    return j

#function to shift Bel probabilities given delta_j
def shiftBelief(Bel,delta_j):
    m= np.size(Bel)
    #copy the Bel to keep an old copy
    Belold = Bel
    Bel = np.zeros(np.size(Bel))
    #now loop through and shift
    print("delta_j = "+str(delta_j))
    print(range(max([0,delta_j]),min(m-delta_j,m)))
    for j in range(max([0,delta_j]),min(m-delta_j,m)):
        # print("j="+str(j))
        Bel[j+delta_j]=Belold[j]
    return Bel

#function to get the x coordinate of most likely j's center
def getEstimate(j,Bel,d):
    m = np.size(Bel)
    x = j*d/m + d/(2*m)
    return x

# Function to display our belief histogram
def dispBel(Bel):
    #create an image to display to show us our belief
    Bel_image = np.array([Bel])
    Bel_image = np.dot(np.ones((np.size(Bel),1)),Bel_image)
    Bel_image = cv2.resize(Bel_image,(3*np.size(Bel),3*np.size(Bel)))
    Bel_image = Bel_image/np.max(Bel)
    cv2.imshow('Belief',Bel_image)

#now put these into a loop.
#define a "Bel" which is actually an array of probabilities
Bel = np.zeros(100)
d = 1 #1cm grid 1m long

#initialize our belief to a particular position
x0 = 0.1
j0 = getCell(x0,Bel,d)

#initialize our estimates.
xhat = x0
jhat = j0
Bel[jhat] = 1.0

#define a histogram H that represents uncertainty in motion model
Hx = np.array([.1,.8,.1])

quit = False
u = 0
j_old = jhat
dispBel(Bel)
while not:
    #get key from openCV. Wait until we have one!!
    key = cv2.waitKey(-1)
    if(key==ord('q')):
        quit=True
        u=0
    elif(key==ord('l')):
        u=.01
    elif(key==ord('j')):
        u = -.01
    elif(key==ord('k')):
        u=0
    # print("xhat-: "+str(xhat)+", u = "+str(u)+", jhat- = "+str(jhat))
    #step 1: propagate estimate using motion model
    xhat = updateMotion(xhat,u)
    # print("after update, xhat = "+str(xhat))
    #step 2: get new jhat
    jhat = getCell(xhat,Bel,d)
    # print("jhat- ="+str(jhat))
    #step 3: calculate shift
    delta_j = jhat-j_old
    # print("jhat = "+str(jhat)+", j_old = "+str(j_old)+", delta = "+str(delta_j))
    #update j old for next time
    j_old = jhat
    #step 4: shift belief!!
    Bel = shiftBelief(Bel,delta_j)
    #step 5: update Belief with uncertainty by convolving.
    Bel = np.convolve(Hx,Bel,'same')
    #get new jhat
    jhat = np.argmax(Bel)
    #get new xhat
    xhat = getEstimate(jhat,Bel,d)
    # print("xhat+: "+str(xhat)+", u = "+str(u)+", jhat+ = "+str(jhat))
    #display Bel for us.
    dispBel(Bel)
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

#function to update dynamics
def updateMotion(x,u):
    #just a simple model for now
    x = x+u
    return x


#function to get the index j in Bel given an x.
def getCell(x,Bel,d):
    m = np.size(Bel)
    j = int((x-d/(2*m))*m/d)
    return j

#function to shift Bel probabilities given delta_j
def shiftBelief(Bel,delta_j):
    m= np.size(Bel)
    #copy the Bel to keep an old copy
    Belold = Bel
    Bel = np.zeros(np.size(Bel))
    #now loop through and shift
    print("delta_j = "+str(delta_j))
    print(range(max([0,delta_j]),min(m-delta_j,m)))
    for j in range(max([0,delta_j]),min(m-delta_j,m)):
        # print("j="+str(j))
        Bel[j+delta_j]=Belold[j]
    return Bel

#function to get the x coordinate of most likely j's center
def getEstimate(j,Bel,d):
    m = np.size(Bel)
    x = j*d/m + d/(2*m)
    return x

#given a cell j, how likely is it that we would get a measurement z?
def getPzx(z,Hz,Bel,d):
    #store the half-size of the measurement histogram
    halfHz = int((np.size(Hz)-1)/2)
    #create a padded version that will handle cells through range of Bel
    Pzx_padded = np.zeros(np.size(Bel)+2*halfHz)
    #now determine which cell the measurement SHOULD be in w/o noise
    zj = getCell(z,Bel,d)
    #now fill in Pzx with this histogram, centered at the right spot
    Pzx_padded[zj-halfHz:zj-halfHz+np.size(Hz)] = Hz
    #now slice the histogram and return
    return Pzx_padded[halfHz-1:np.size(Bel)]


def bayesUpdate(Bel,Hz,z,d):
    #fill in probability of getting this measurement given z
    Pzx = getPzx(z,Hz,Bel,d)
    #initialize Pz
    Pz = 0
    #now loop through all cells and apply Bayes's rule
    for j in range(0,np.size(Bel)):
        Bel[j] = Bel[j]*Pzx[j]
        Pz+=Bel[j]
    print("Pz = "+str(Pz))
    #now loop through and normalize using Pz
    for j in range(0,np.size(Bel)):
        Bel[j]= Bel[j]/Pz
    return Bel


# Function to display our belief histogram
def dispBel(Bel,xtrue,d):
    #create an image to display to show us our belief
    Bel_image = np.array([Bel],dtype=np.float32)
    Bel_image = np.dot(np.ones((np.size(Bel),1),dtype=np.float32),Bel_image)
    Bel_image = cv2.cvtColor(Bel_image, cv2.COLOR_GRAY2BGR)
    jtrue = getCell(xtrue,Bel,d)
    Bel_image = Bel_image/np.max(Bel)
    Bel_image[:,jtrue,:] = (0,0,255)
    Bel_image = cv2.resize(Bel_image,(3*np.size(Bel),3*np.size(Bel)))
    #Bel_image = Bel_image/np.max(Bel)

    cv2.imshow('Belief',Bel_image)

#now put these into a loop.
#define a "Bel" which is actually an array of probabilities
Bel = np.zeros(100)
d = 1 #1cm grid 1m long

#initialize our belief to a particular position
x0 = 0.1
j0 = getCell(x0,Bel,d)

#initialize our estimates.
xhat = x0
xtrue = x0
jhat = j0
Bel[jhat] = 1.0

#define a histogram Hx that represents uncertainty in motion model
Hx = np.array([.1,.8,.1])
#Define histogram Hz representing measurement noise
Hz = np.array([.025, .95, .025])

quit = False
u = 0
j_old = jhat
dispBel(Bel,xtrue,d)

#normalizer (P(z))
zeta = 0

measurement_interval = 10
measurement_counter = 0

while not:
    #get key from openCV. Wait until we have one!!
    key = cv2.waitKey(-1)
    if(key==ord('q')):
        quit=True
        u=0
    elif(key==ord('l')):
        u=.01
    elif(key==ord('j')):
        u = -.01
    elif(key==ord('k')):
        u=0
    # print("xhat-: "+str(xhat)+", u = "+str(u)+", jhat- = "+str(jhat))
    #step 1: propagate estimate using motion model
    xhat = updateMotion(xhat,u)
    #keep track of "true" x as well
    xtrue = updateMotion(xtrue,u)
    # print("after update, xhat = "+str(xhat))
    #step 2: get new jhat
    jhat = getCell(xhat,Bel,d)
    # print("jhat- ="+str(jhat))
    #step 3: calculate shift
    delta_j = jhat-j_old
    # print("jhat = "+str(jhat)+", j_old = "+str(j_old)+", delta = "+str(delta_j))
    #update j old for next time
    j_old = jhat
    #step 4: shift belief!!
    Bel = shiftBelief(Bel,delta_j)
    #step 5: update Belief with uncertainty by convolving.
    Bel = np.convolve(Hx,Bel,'same')

    measurement_counter+=1
    #step 6: update belief based on new measurement if we have one!
    if measurement_counter==measurement_interval:
        measurement_counter=0
        #create a fake measurement-- assume came from a noisy dist sensor
        #for now, assume Gaussian noise w/ std of .025cm. This means
        #that roughly, 95% conf is within a cell (4 std total, 2 each way).
        z = xtrue + .00025*np.random.randn()
        #now pass z and Hz to bayes update function
        Bel = bayesUpdate(Bel,Hz,z,d)


    #get new jhat
    jhat = np.argmax(Bel)
    #get new xhat
    xhat = getEstimate(jhat,Bel,d)

    # print("xhat+: "+str(xhat)+", u = "+str(u)+", jhat+ = "+str(jhat))
    #display Bel for us.
    dispBel(Bel,xtrue,d)
````
