## How can we characterize a first-order response's  *transient behavior?*

Given a linear differential equation of the form $a_1\dot{y} + a_0y = b_0U$, we can use the *time constant* to characterize its **transient behavior.**

### Time Constant
* For a first order system, the system's single eigenvalue is $p=-\frac{a_0}{a_1}$.  
* We define the time constant to be $\tau = \frac{a_1}{a_0} = -\frac{1}{p}$
**WHY?**
When $t = \tau$, then the differential equation's **homogeneous** solution is
\begin{equation}
y_h(t) = y_0e^{-\frac{t}{\tau}} = y_oe^{-1} = 0.367879y_0
\end{equation}

This means that if $y(\tau) = 0.368 y_0$, then approximately 63.2% of the **change** from $t_0$ to $t_{ss}$ has occurred when $t=\tau$. For step responses, this means that the time constant can be pulled off of a plot by finding the time $\tau$ at which $y(t) = 0.632(y_{ss}-y_0)$. 

## How can we characterize a first-order response's *steady state* behavior?

## How can we characterize a first-order response's *steady state* behavior?

### Steady State Gain $\equiv \frac{\Delta y_{ss}}{\Delta U_{ss}}$

Given a first-order linear differential equation of the form $a_1\dot{y} + a_0y = b_0U$, setting $\dot{y}=0$ to indicate *steady state*, and solving for the change in our output at steady state yields

$$\Delta y_{ss} = y_{ss}-y_0 = \left(U_{ss} - U_0\right)\frac{b_0}{a_0}$$

This allows us to write an equation for "steady state gain" using the first-order differential equation's $a_0$ and $b_0$ terms.

$$K_{ss} = \frac{y_{ss}-y_0}{U-U_0} = \frac{\Delta y_{ss}}{\Delta u_{ss}} = \frac{b_0}{a_0} $$
