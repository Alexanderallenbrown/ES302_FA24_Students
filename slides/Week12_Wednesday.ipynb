## Lit Review Feedback Themes
1. If a sentence has no informational content, it should not be in your document. 
> These terms were then developed and interchanged until they became an accurate representation of the research questions proposed. 
2. If a statement is *falsifiable* (if it is a claim that could be either true or false), it needs evidentiary support in the form of a citation, a figure, or a table
> Snake robots are often perceived as frightening, even though they are useful in many household tasks
3. Introspective statements about the assignment, *especially* those that break the "4th wall" are not appropriate for a technical document and should not be used.
> I chose to study snake robots because I hoped to learn more about legless locomotion. I hoped that it wouldn't be too difficult but also challenging enough to build into a final project.
