### Last class we derived the forward differential kinematics for ArtyBot

Then, the forward differential kinematic equation for the Artybot can be written:

\begin{equation}
\begin{bmatrix} ^A\dot{x}_p \\ ^A\dot{y}_p \end{bmatrix} = 
\begin{bmatrix} \frac{\partial ^Ax_p}{\partial \theta_2} & \frac{\partial ^Ax_p}{\partial \theta_3} \\ \frac{\partial ^Ay_p}{\partial \theta_2} & \frac{\partial ^Ay_p}{\partial \theta_3} \end{bmatrix} 
\begin{bmatrix} \dot{\theta}_2 \\ \dot{\theta_3}\end{bmatrix}
\end{equation}

Where the terms in the Jacobian were given by:
\begin{equation}
\begin{aligned}
\frac{\partial ^Ax_p}{\partial \theta_2} &= - l_2\sin\theta_2 + l_3\left(-\sin\theta_2\cos\theta_3 - \cos\theta_2\sin\theta_3\right) \\
\frac{\partial ^Ay_p}{\partial \theta_2} &= l_2\cos\theta_2 + l_3\left(\cos\theta_2\cos\theta_3 - \sin\theta_2\sin\theta_3\right) \\
\frac{\partial ^Ax_p}{\partial \theta_3} &=  -l_3\left(\cos\theta_2\sin\theta_3 + \sin\theta_2\cos\theta_3\right) \\
\frac{\partial ^Ay_p}{\partial \theta_3} &=  l_3\left(-\sin\theta_2\sin\theta_3 + \cos\theta_2\cos\theta_3\right) 
\end{aligned}
\end{equation}

## Can we derive the same Jacobian using velocity equations directly?

A rigid body velocity analysis yields a complete description of $^A\vec{v}_P$ as a function of the robot's configuration and its configuration velocity:

\begin{equation}
^A\vec{v}_P = l_2\dot{\theta_2}\left(\cos\theta_2\hat{y}_A-\sin\theta_2\hat{x}_A\right) + l_3\left(\dot{\theta_2}+\dot{\theta_3}\right)\left(\cos\left(\theta_2+\theta_3\right)\hat{y}_A-\sin\left(\theta_2+\theta_3\right)\hat{x}_A\right)
\end{equation}

Writing Equation 16 as a vector-matrix equation for $^A\vec{v}_P = \begin{bmatrix}^A\dot{x}_P & ^A\dot{y}_P\end{bmatrix}^T$ yields:

\begin{equation}
\begin{bmatrix} ^A\dot{x}_p \\ ^A\dot{y}_p \end{bmatrix} = 
\begin{bmatrix} -\left(l_2\sin\theta_2+l_3\sin\left(\theta_2+\theta_3\right)\right)\ & -l_3\sin\left(\theta_2+\theta_3\right) \\
l_2\cos\theta_2 +l_3\cos\left(\theta_2+\theta_3\right) & l_3\cos\left(\theta_2+\theta_3\right) \end{bmatrix} 
\begin{bmatrix} \dot{\theta}_2 \\ \dot{\theta_3}\end{bmatrix}
\end{equation}

## Can we derive the same Jacobian using velocity equations directly?

**If this equation is consistent with our definition of the Jacobian and its role in differential kinematics, Equation 17 should be equivalent to Equation 9 with Equation 8 substituted in! In other words, the following terms should be equivalent:**

\begin{equation}
\begin{aligned}
\frac{\partial ^Ax_p}{\partial \theta_2} &=& - l_2\sin\theta_2 + l_3\left(-\sin\theta_2\cos\theta_3 - \cos\theta_2\sin\theta_3\right) &=&  -\left(l_2\sin\theta_2+l_3\sin\left(\theta_2+\theta_3\right)\right) \\
\frac{\partial ^Ay_p}{\partial \theta_2} &=& l_2\cos\theta_2 + l_3\left(\cos\theta_2\cos\theta_3 - \sin\theta_2\sin\theta_3\right) &=& l_2\cos\theta_2 +l_3\cos\left(\theta_2+\theta_3\right) \\
\frac{\partial ^Ax_p}{\partial \theta_3} &=&  -l_3\left(\cos\theta_2\sin\theta_3 + \sin\theta_2\cos\theta_3\right) &=& -l_3\sin\left(\theta_2+\theta_3\right) \\
\frac{\partial ^Ay_p}{\partial \theta_3} &=&  l_3\left(-\sin\theta_2\sin\theta_3 + \cos\theta_2\cos\theta_3\right) &=& l_3\cos\left(\theta_2+\theta_3\right)
\end{aligned}
\end{equation}

Using the angle addition identities:
\begin{equation}
\begin{aligned}
\sin\left(\theta_2+\theta_3\right) &= \sin\theta_2\cos\theta_3 +\cos\theta_2\sin\theta_3 \\
\cos\left(\theta_2+\theta_3\right) &= \cos\theta_2\cos\theta_3 -\sin\theta_2\sin\theta_3 
\end{aligned}
\end{equation}
We can see that they are! This confirms that the Jacobian can be obtained from a positional kinematic solution, but also from a direct analysis of the velocities of a robot's links.

## Using Differential Kinematics to compute *static* robot joint forces and torques

One of the most powerful ways to use differential kinematics is for an analysis of the *static* forces and moments in a robot in configuration space, given an externally applied force and/or moment in task space. We might ask, for example, what motor torques $\mu_2$ and $\mu_3$ are required on the ArtyBot if someone pushes on its pen $P$ with a force $\vec{F}_{P,applied}$. 

Note that $\mu$ will be used to denote Torque rather than $T$ to avoid confusion with our homogeneous transformations $T$.

## Wait... have you learned this before?

In a static force and moment analysis we assume that a rigid body or system of rigid bodies (truss, rigid body, whatever) is *not accelerating*. This means that we can write two key vector equations:

\begin{equation}
\begin{aligned}
\sum\vec{F} &= \vec{0} \\
\sum\vec{\mu} &= \vec{0}
\end{aligned}
\end{equation}

.

.

.

## Using differential kinematics to solve statics problems

We can define a "generalized force" called a "Wrench" as:

\begin{equation}
\vec{\tau} = \begin{bmatrix} F_x & F_y & F_z & \mu_x & \mu_y & \mu_z \end{bmatrix}^T
\end{equation}

For a robot with $N$ links and $N-1$ joints connecting them (in an open kinematic chain), we can say that a Wrench $\tau_i$ exists for each of the $N-1$ joints that represents how much force and torque must be reacted by that joint. The wrench will be partially reacted by the structure of the joint (reaction forces at the 'pin' of a hinge joint, for example) and partially reacted by the motor that actuates the joint. 

For example, on the ArtyBot, both of the planar joints actuating $\theta_2,\theta_3$ are revolutes about the $\hat{z}$ direction, so the torque reacted by joint 2 would be the $\mu_z$ portion of $\vec{\tau}_2$.

## Using differential kinematics to solve statics problems

Looking at the general case for a robot with a Wrench applied to its end effector, let's call that wrench $\vec{\tau}_{P,e}$. This is the wrench that the **environment** is applying to the robot. Static equivalency (see section 4.1) insists that by Newton's 3rd Law:
\begin{equation}
\vec{\tau}_{P,e}=-\vec{\tau}_{P}
\end{equation}

where $\vec{\tau}_{P}$ is the wrench the **robot applies to the environment**.

## Using differential kinematics to solve statics problems

In ES103, you learned that **for an idealized transducer**,

\begin{equation}
\dot{W}_{in}=\dot{W}_{out}
\end{equation}

For a rotational to mechanical transducer, for example, you could write (in vector form):

\begin{equation}
\vec{\mu}\cdot\vec{\Omega} = \vec{F}\cdot\vec{v}
\end{equation}

**in a more general sense**, we could say that a robot is a little bit like a "transducer" from the configuration space to the task space. In other words, the power $\dot{W}_{conf}$ the robot **supplies in the configuration space** must be equal to the power $\dot{W}_{task}$ that the robot **generates in the task space**. 

## Using differential kinematics to solve statics problems

We can express the total power (as a sum of both translational and rotational)  supplied by the robot in configuration space as a dot product between the **wrench** representing the "generalized forces" applied by a robot's actuators and the **twist** representing the robot's joints' generalized velocities:
\begin{equation}
\dot{W}_{conf} = \vec{\tau}_P^T\dot{q} = \begin{bmatrix}\tau_1 & \tau_2 & \cdots & \tau_{N-1} \end{bmatrix}\begin{bmatrix}\dot{q}_1 \\ \dot{q}_2 \\ \vdots \\ \dot{q}_{N-1}\end{bmatrix}
\end{equation}

We can similarly express the total power (as a sum of both translational and rotational) required by the environment in task space as another dot product:
\begin{equation}
\dot{W}_{task} = \vec{\tau}_{P,e}^T\vec{\mathcal{V}} = \begin{bmatrix}F_x & F_y & F_z & \mu_x & \mu_y & \mu_z \end{bmatrix}\begin{bmatrix}v_x \\ v_y \\ v_z \\ \omega_x \\ \omega_y \\ \omega_z\end{bmatrix}
\end{equation}

And we know that if there are no losses in the robot, if it acts as an "idealized transducer," then:
\begin{equation}
\dot{W}_{conf} = \dot{W}_{task}
\end{equation}

## Using differential kinematics to solve statics problems

Luckily, we know from differential kinematics that:
\begin{equation}
\vec{\mathcal{V}} = J(\vec{q})\dot{\vec{q}}
\end{equation}

which means that we can write our first law equivalence of input and output power as:

\begin{equation}
\vec{\tau}_c^T\dot{\vec{q}} = \vec{\tau}_{P}^T\vec{\mathcal{V}} = \vec{\tau}_{P}^TJ(\vec{q})\dot{\vec{q}}
\end{equation}

Canceling the common power of $\dot{\vec{q}}$ (a very common move in ES103!) yields:

\begin{equation}
\vec{\tau}_c^T = \vec{\tau}_{P}^TJ(\vec{q})
\end{equation}

Which implies:

\begin{equation}
\vec{\tau}_c = J(\vec{q})^T\vec{\tau}_{P}
\end{equation}

Where, as a reminder, $\vec{\tau}_P$ is the vector of generalized joint forces applied by the robot in *configuration space* while $\vec{\tau}_{P,e}$ is the vector of applied generalized end-effector forces *in task space*. 

## Using differential kinematics to solve statics problems

### Summary

The robot's Jacobian can be used to map a "wrench" applied by the robot's joints in configuration space to the "wrench" applied by the robot in task space. Namely:

\begin{equation}
\vec{\tau}_c = J^T\vec{\tau}_P = \begin{bmatrix}\frac{\partial x}{\partial q_1} & \frac{\partial y}{\partial q_1} & \cdots & \frac{\partial \theta_z}{\partial q_1} \\ \frac{\partial x}{\partial q_2} & \frac{\partial y}{\partial q_2} & \cdots & \frac{\partial \theta_z}{\partial q_2} \\ \cdots & \cdots & \cdots & \cdots \\ \frac{\partial x}{\partial q_n} & \frac{\partial y}{\partial q_n} & \cdots & \frac{\partial \theta_z}{\partial q_n} \end{bmatrix}\begin{bmatrix}F_x \\ F_y \\ F_z \\ \vdots \\ \mu_z\end{bmatrix}
\end{equation}

The same approach can be used to invert the process using the pseudoinverse of the Jacobian's transpose. In other words,

\begin{equation}
\vec{\tau}_P = \left(J^T\right)^+\vec{\tau}_c
\end{equation}

**NOTES**
* When a "singularity" occurs in the Jacobian and it is not invertible, this affects *both* the static force/moment analysis as well as the velocity analysis, but in different ways.
* Be careul about what coordinate system you use when deriving "J" and defining the "task space." Know which coordinate system you're expressing $\vec{\tau}_{P}$ in!

## Exercise: Make Romi move with a specified forward velocity and yaw rate. 

If we wanted Romi to move in a circle of radius $R$ at some foward speed $U$, we could compute a goal yaw rate for the robot as:

\begin{equation}
\psi_g = \frac{U}{R}
\end{equation}

This means that we could specify a "goal" for the Romi's body-fixed task space as:

\begin{equation}
\vec{q}_g = \begin{bmatrix}^B\dot{x}_g & ^B\dot{y}_g & ^B\dot{\psi}_g\end{bmatrix}=\begin{bmatrix}U & 0 &  \frac{U}{R}\end{bmatrix}
\end{equation}

Then, we could make the Romi achieve this goal in its "body-fixed task space" by moving its wheels at:

\begin{equation}
\vec{q} = \begin{bmatrix}\omega_R \\ \omega_L\end{bmatrix} = \underbrace{\begin{bmatrix} \frac{R_w}{2} & -\frac{R_w}{2} \\ 0 & 0 \\\frac{R_w}{2r} & -\frac{R_w}{2r} \end{bmatrix}^+}_{J^+}\underbrace{\begin{bmatrix}U  \\ 0  \\  \frac{U}{R}\end{bmatrix}}_{\vec{\mathcal{V}}_b}
\end{equation}
