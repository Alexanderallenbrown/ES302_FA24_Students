## Representing $^A\vec{F}$

We learned about how to tranform vectors from one frame to another using *rotations*, but often, a *translation* from one frame's origin to the other frame's origin is also required. To write $^A\vec{F}$, we must write both the offset between frames $A$ and $B$ *and* the rotation of $^B\vec{f}$ into $A$'s frame.

\begin{equation}
^A\vec{F} = ^A\vec{P} + ^A_BR_z(\theta)^B\vec{f}
\end{equation}

## We can formalize this process for transforming between frames with some easy notation.

### Notation for expressing one frame with respect to another

If we want to express this compactly, we can say that the frame $B$ can be expressed relative to (rather from the perspective of) frame $A$ by expressing the translation and rotation as a set:

\begin{equation}
\{B\} = \left\{ _B^AR, ^AP \right\}
\end{equation}

Where, in general, $^AP$ represents the vector from the *origin* of frame $A$ to the *origin* of frame $B$ expressed in $A$'s coordinates, and the rotation vector $_B^AR$ represents the rotation *from* $A$ *to* $B$ (enabling us to write $B$'s unit vectors in terms of $A$'s unit vectors).


## We can use this mapping to express any position vector $^B\vec{Q}$ in frame A as $^A\vec{Q}$

\begin{equation}
^A\vec{Q} = ^A\vec{P} + ^A_B R ^B\vec{Q}
\end{equation}

### This can be written very compactly if we use "Homogeneous Coordinates"

\begin{equation}
\underbrace{\begin{bmatrix}^A\vec{P} \\ 1\end{bmatrix}}_{^A\vec{P}_h}
=
\underbrace{\left[ \begin{array}{cc|c}
_B^AR & & ^A\vec{P} \\ \hline
0& 0& 1
\end{array}\right]}_{_B^A T} \underbrace{\begin{bmatrix}^B\vec{P} \\ 1\end{bmatrix}}_{^B\vec{P}_h}
\end{equation}

This allows the entire transformation to be written simply as:

\begin{equation}
^A\vec{P}_h = _B^A T ^B\vec{P}_h
\end{equation}

### Inverse Transformations

Due to the properties of a homogeneous transformation matrix, the inverse transformation matrix is easy to compute. There's no need to take the general inverse of a 3x3 (for a 2D homogeneous transformation) or 4x4 matrix (for a 3D homogeneous transformation) by hand.

\begin{equation}
_A^B T = _B^A T^{-1} = \left[ \begin{array}{cc|c} _B^A R^T & &- ^A_B R^T\:  ^A\vec{P} \\ \hline 0& 0& 1 \end{array}\right]
\end{equation}

### Multiple Transformations
Note that in many cases, a robot may have many frames, perhaps attached to different "links" in a robot arm. This notation for transformations allows us to *concatenate* multiple transformations by simply multiplying each successive transformation matrix together. For example, if we had a third frame on the robot, we could write:

\begin{equation}
_C^A T = _B^A T\:_C^B T
\end{equation}

Think of the superscripts and subscripts similarly to units in a dimensional analysis-- the superscripts and subscripts will "cancel" during the intermediate transformation between frames C and B to leave the "final" transformation from C to A.

### Transformations in Python

Using numpy, it is relatively simple to write functions that perform coordinate transformations for us. We do this by constructing a rotation matrix first (using the function we developed above), and then building the transformation matrix. We can also write functions to actually *do* the transformation (and the inverse transformation) for us. See below.

```python
import numpy as np
def create2DRotationAB(theta):
    """ Create a 2D rotation matrix AB_R (R from A to B ) to express the unit vectors of coordinate system B
        in terms of the unit vectors of coordinate system B.
        If you wish to do the opposite (transform unit vectors of A into B's coordinates), you can invert/transpose this matrix.
        The angle you feed this function should be in RADIANS.

        note that the matrix equation here is [xhatb,yhatb]' = R*[xhata,yhata]' where ' represents transpose.
    """
    return np.array([[np.cos(theta),np.sin(theta)],[-np.sin(theta),np.cos(theta)]])

def create2DTransform(aX_b,aY_b,theta):
    """ Create a 2D transform matrix that takes any point in coordinate system B and expresses it in A's coordinate system.
        This performs a translation and a rotation assuming that coordinate system B's position in A's coordinates is known,
        and the angle theta by which coordinate system B is rotated relative to A is known.
    """
    #create rotation matrix
    R = create2DRotationAB(theta).T #because we wrote our rotation function from A to B, need the transpose to get B to A.
    #create vector expressing B's position relative to A's origin in A coordinates
    aPb = np.array([[aX_b],[aY_b]])
    #stack these two horizontally
    baT = np.hstack((R,aPb))
    #now stack this vertically with a [0,0,1] to turn into homogeneous transformation
    baT = np.vstack((baT,np.array([0,0,1])))
    #now return this transformation
    return baT

def transform2DBA(bXq,bYq,T):
    """transform a point q expressed in coordinate system B's frame into coordinate system A's frame, where
        T represents the homogeneous transformation from B coords to A coords, assuming that coordinate frame
        B is rotated by a known angle theta relative to A, and is located at a known offset in A's frame.
    """
    result = np.dot(T,np.array([[bXq],[bYq],[1]]))
    #returns aXq,aYq (q's X and Y coordinate in A's frame)
    return result[0][0],result[1][0]

def transform2DAB(aXq,aYq,T):
    """transform a point q expressed in coordinate system A's frame into coordinate system B's frame, where
        where T represents the homogeneous transformation from B coords to A coords, assuming that coordinate frame
        B is rotated by a known angle theta relative to A, and is located at a known offset in A's frame.
        This function uses the INVERSE of T to accomplish this.
    """
    result = np.dot(np.linalg.inv(T),np.array([[aXq],[aYq],[1]]))
    #returns bXq,bYq (q's X and Y coordinate in B's frame)
    return result[0][0],result[1][0]
```

### Example-- Transforming local sensor measurements on a robot to a global frame

Consider the example from Figure 1. We will say that the robot is located at $^A \vec{P} = \begin{bmatrix} 4 \\ 3\end{bmatrix}$ meters from a global origin. Let's imagine that the robot, whose body-fixed frame $B$ is oriented at a "yaw" angle $\theta=0.5$ radians with respect to the earth-fixed frame $A$. Let's further assume that the robot measures the "apple" using a [LIDAR sensor](https://en.wikipedia.org/wiki/Lidar) to be at $^B\vec{f} = \begin{bmatrix} 0.5 \\ 2 \end{bmatrix}$ m with respect to the robot's frame.

Let's set up this transformation using our utility functions and see if we can correctly compute where the robot might store the apple's position on a "map" that is expressed in the allocentrc, earth-fixed frame $A$. This amounts to performing the transformation:

\begin{equation}
^A\vec{F}_h = ^A_BT\: ^B\vec{f}_h
\end{equation}

The situation is plotted to scale in the figure below.

````python
from matplotlib.pyplot import *
rcParams['text.usetex'] = False

def drawAxis(myorigin,rotation,mycolor,mylabel):
    arrow(myorigin[0],myorigin[1],rotation[0,0],rotation[0,1],color=mycolor,width=0.01,label=mylabel,length_includes_head=True)
    arrow(myorigin[0],myorigin[1],rotation[1,0],rotation[1,1],color=mycolor,width=0.01,length_includes_head=True)
    text(myorigin[0]+rotation[0,0],myorigin[1]+rotation[0,1],r'$\hat{x}$',color=mycolor)
    text(myorigin[0]+rotation[1,0],myorigin[1]+rotation[1,1],r'$\hat{y}$',color=mycolor)

def drawVector(myorigin,rotation,howlong,mycolor,mylabel,mywidth=0.015):
    arrow(myorigin[0],myorigin[1],howlong*rotation[0,0],howlong*rotation[0,1],color=mycolor,width=mywidth,label=mylabel,length_includes_head=True)
    text(myorigin[0]+howlong*rotation[0,0],myorigin[1]+howlong*rotation[0,1],mylabel,color=mycolor)

AB_R = create2DRotationAB(0.5)
origin=np.array([4,3])
rotation2 = create2DRotationAB(0.5+np.arctan2(2,.5))

figure()
drawVector(origin,rotation2,2.06,'r',r'$^B\vec{f}$')
drawVector(np.array([0,0]),create2DRotationAB(np.arctan2(3,4)),5,'m',r'$^A\vec{P}$')
drawVector(np.array([0,0]),create2DRotationAB(np.arctan2(5,3.5)),6.103,'g',r'$^A\vec{F}$')

drawAxis(np.array([0,0]),np.eye(2),'k','frame A (global)')
drawAxis(origin,AB_R,'b','frame B (robot-fixed)')
axis('equal')
legend()
grid('minor')
show()
````

Remember, Euler integration allows us to compute a "new" value for some integral $y(k)$ based on an "old" value $y(k-1)$, a velocity $\dot{y}(k-1)$, and a timestep $\Delta t$ by:

\begin{equation}
y(k) = y(k-1) + \dot{y}(k-1)\Delta t
\end{equation}

Which can be written without using indices and without storing our estimates in a "vector" by simply writing:

\begin{equation}
y_{new} = y_{old} + \dot{y}_{old}\Delta t
\end{equation}
